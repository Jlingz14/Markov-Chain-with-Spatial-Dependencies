```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_excel('covid_first_wave.xlsx')
df.tail(10)
```

```python
# DÃ©finir la rÃ©gion comme l'index
df.set_index('Region', inplace=True)

# DÃ©terminer le nombre de sous-graphiques
num_regions = len(df)
num_cols = 4
num_rows = math.ceil(num_regions / num_cols)

# CrÃ©er des sous-graphiques
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15,10), sharex=True)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Tracer les donnÃ©es pour chaque rÃ©gion
for i, region in enumerate(df.index):
    df.loc[region].plot(kind='line',marker='.', ax=axes[i], title=f'Data for the Region {region}')
    axes[i].set_xlabel('Dated')
    axes[i].set_ylabel('New case')
    axes[i].grid(False)

# Supprimer les sous-graphiques vides
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Ajuster les espacements
plt.tight_layout()

# Afficher la figure
plt.savefig("Covid_per.jpg",dpi=350)
plt.show()
```
